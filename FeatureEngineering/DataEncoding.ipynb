### Nominal/OHE ENCODING
- Each Unique Categorical Value is convereted into a feature.
- Red Green Blue are categorical Feature converting into Numerical Feature

## Disadvatanges
- More Categories, More FEATURES
- Sparse Matrix ----> OverFitting.

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
### Create a DataFrame
df = pd.DataFrame({
    "color":['red','blue','green','green','blue','red']
})
```

```python
encoder = OneHotEncoder()
```

```python
## Perform Fit and Transform
encoded = encoder.fit_transform(df[['color']]).toarray()
```

```python
encoded_df = pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
encoded_df
```

```python
encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df,encoded_df],axis=1)
```

```python
##PRACTICE

import seaborn as sns
import numpy as np

df=sns.load_dataset('tips')
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
gender = encoder.fit_transform(df[['sex']]).toarray()
gendercols = encoder.get_feature_names_out()
timeofday = encoder.fit_transform(df[['day']]).toarray()
timecols = encoder.get_feature_names_out()
times = encoder.fit_transform(df[['time']]).toarray()
timecols2 = encoder.get_feature_names_out()
smoker = encoder.fit_transform(df[['smoker']]).toarray()
smokercol=encoder.get_feature_names_out()

df2 = pd.DataFrame(np.hstack([gender,timeofday,times,smoker]),
                  columns = list(gendercols) + list(timecols) + list(timecols2) + list(smokercol)
)
```

```python
df2
```

```python
pd.concat([df,df2],axis=1)
```

### Label Encoding
- Assign Unique Value to each categories

Disadvantages
- Assigning Label creates rank that effect the ML model

```python
from sklearn.preprocessing import LabelEncoder
### Create a DataFrame
df = pd.DataFrame({
    "color":['red','blue','green','green','blue','red']
})

labelencoder = LabelEncoder()
labelencoder.fit_transform(df[['color']])
```

```python
labelencoder.transform([['green']])
```

### Ordinal Encoding
- Categorical Data that requires ranking

for example Education Completed. Bachelors,College,Graduate,PHD

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
df = pd.DataFrame({
    'size':['small','medium','large','medium','small','large']
})
```

```python
### Create a Instance of OrdinalEncoder and then fit_transform
encoder = OrdinalEncoder(categories=[['small','medium','large']]) ## Rank 0 --- > Small, 1--> Medium, 2--->Large
```

```python
encoder.fit_transform(df[['size']])
```

### Target Guided Ordinal Encoding
- Converting Categorical Feature into numerical by taking the mean of the of the target belonging to the same categorical Feature

```python
df = pd.DataFrame({
    'city':['NewYork','Japan','Italy','NewYork','Pakistan','Iran'] ,
    'price':[200,150,300,250,180,320]
})
```

```python
mean_price=df.groupby('city')['price'].mean().to_dict()
df['cityencoded']=df['city'].map(mean_price)
```
