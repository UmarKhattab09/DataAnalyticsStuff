### FLIGHT DATASET EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('flight_price.xlsx')
```

```python
### Convert The Date of Journey into Date
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
df
```

```python
df.info() # --- > Still In Object
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.dtypes
```

```python
### Dropping Date of Journey Column
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
#----> Arrival Time, Split
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
df['ArrivalHour'] = df['Arrival_Time'].str.split(':').str[0]
df['ArrivalMin'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head(2)
```

```python
df['ArrivalHour']=df['ArrivalHour'].astype(int)
df['ArrivalMin']=df['ArrivalMin'].astype(int)
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df['DeptHour'] = df['Dep_Time'].str.split(':').str[0]
df['DeptMin'] = df['Dep_Time'].str.split(':').str[1]
df['DeptHour']=df['DeptHour'].astype(int)
df['DeptMin']=df['DeptMin'].astype(int)
df.head(2)
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
#---> Total Stops
df['Total_Stops'].unique()

```

```python
df[df['Total_Stops'].isna()]
###---> can be
modevalue = df[df['Total_Stops'].notna()]['Total_Stops'].mode()[0]
modevalue
```

```python
df['Total_Stops']=df['Total_Stops'].fillna(modevalue)
df['Total_Stops'].isna().sum()
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder(categories=[['non-stop','1 stop','2 stops','3 stops','4 stops']]) ## Rank 0 --- > Small, 1--> Medium, 2--->Large
encoder.fit_transform(df[['Total_Stops']])
```

```python
df['TotalStopsEncoded'] = encoder.fit_transform(df[['Total_Stops']])
df.head()
```

```python
df.drop(['Route','Total_Stops'],axis=1,inplace=True)
```

```python
df.head()
```

```python
# ---- > AirLine, Source, Destination , Additional Info ----> ONE HOT ENCODING
# ----> Duration ----> Lambda Split.
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoded = encoder.fit_transform(df[['Airline','Source','Destination']]).toarray()
encoded_df = pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
encoded_df
```

```python
df_copy=pd.concat([df,encoded_df],axis=1)
df_copy.drop(['Airline','Source','Destination'],axis=1,inplace=True)

df_copy.info()
```
